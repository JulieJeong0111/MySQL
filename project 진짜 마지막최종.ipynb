```python
import pymysql 


def conn_on():
    
    global conn
    conn = pymysql.connect(
    host='localhost',
    port=3306,
    user='root',
    passwd='root',
    db='phonestore')
```

```python
#############
#항목선택 함수
#############

def 항목선택():
    print()
    print('휴대폰 판매 통합 시스템')
    print('*'*20)
    print('1. 등록')
    print('2. 검색')
    print('3. 종료')
    print()

    choi_num = int(input('원하시는 항목을 선택하세요. : '))
             
    while choi_num < 1 or choi_num > 3 :
        choi_num = int(input('항목 번호를 확인해주세요. : '))
    
    return choi_num

```

```python
############
# 1. 등록 함수
############
def 등록():
    제품명=input('제품명을 입력하세요 : ')
    #제품코드=input('제품코드를 입력하세요 : ')
    출시일=input('출시일을 입력하세요 : ')
    #입고일=input('입고일을 입력하세요 : ')
    제조사=input('제조사를 입력하세요 : ')
    보증기간=int(input('보증기간을 입력하세요 : '))
    용량=input('용량을 입력하세요 : ')
    색상=input('색상을 입력하세요 : ') 
    가격=int(input('가격을 입력하세요 : '))
    화면크기=input('화면크기를 입력하세요 : ')
    무게=input('무게를 입력하세요 : ') 

    conn_on()
    
    query='''
    insert into 상세정보 values(%s,default,%s,%s,%s,%s,%s,%s,%s,%s)
    '''
    cur = conn.cursor()
    cur.execute(query,(제품명,출시일,제조사,보증기간,용량,색상,가격,화면크기,무게))
    
    conn.commit()
    cur.close()
    conn.close()

    answer = input('추가로 등록하시겠습니까? y/n : ')
    
    if answer == 'Y' or answer == 'y' :
        등록()
        
    else :
        global choi_num
        choi_num = choi_num - 1
    
        
    
```

```python
############
# 2.검색 함수
############
def 검색() : 
    #검색
    
    conn_on()
    
    PhoneName = input('기종을 입력하세요.')
    X= PhoneName + '%'
    sqlx = "select count(*) from 상세정보 where 제품명 like '{}'".format(X)
    cur = conn.cursor()    
    cur.execute(sqlx)
    result = cur.fetchall()
    print(PhoneName, '는(은)', result[0][0], '개 있습니다.') 

    question = input('상세보기 하시겠습니까?(Y/N)')

    if question == 'Y' or question == 'y':
        sql_detail = "select * from 상세정보 where 제품명 like '{}'".format(X)    
        cur.execute(sql_detail)
        result1 = cur.fetchall()
        
#         print(result1)

        list_result1=[list(result1[x]) for x in range(len(result1))]

        list_result1_title=['제품명', '제품코드','출시일','제조사','보증기간','용량','색상','가격','화면크기','무게']
        for n in list_result1_title:
            print(n.center(10, ' '), end='')
        print()
        print('*'*130)
           
        cnt=len(list_result1)
        a=0
        while a<cnt:
            b=0
            while b<10:
#                 print(list_result1[a][b],' '*7,end='')
                print("{:^12}".format(list_result1[a][b]),end='')
                b+=1
            a+=1
            print('\n')
            
        
        
        sql_price = "select distinct(제품명), 할인율 from 할인율 where 제품명 like '{}'".format(X)      
        cur.execute(sql_price)
        price_result = cur.fetchall()
#         print(price_result)
        
        list_price_result=[list(price_result[x]) for x in range(len(price_result))]
        print()
        list_sql_price_title=['제품명', '할인율']
        for i in list_sql_price_title:
            print(i.center(7, ' '), end='')
        print()
        print('*'*20)
        
        cnt1=len(list_price_result)
        c=0
        while c<cnt1:
            d=0
            while d<2:
                print(list_price_result[c][d],' '*3,end='')
                d+=1
            c+=1
            print('\n')
        
        print()
        
        cur.close()
        conn.close()
        price = int(input('가격'))
        sale_rate = float(input('할인율'))
        print('행사가는 ' + str(int(price * (1 - sale_rate))) + '원입니다.')
        
    else :
        print('끝')
        
        conn.close()
        cur.close()
        

    
```

```python
#############
# 메인 실행 
#############

choi_num = 0
quit = 3

while choi_num != quit:
    
    choi_num=항목선택()
    
    if choi_num == 1:
        등록()
    elif choi_num == 2:
        검색()
    elif choi_num == 3:
        quit
```
